```python
import pandas as pd
import json
from yelpapi import YelpAPI
import pprint
import config

yelp_api = config.YelpAPI


file = "uscitiesv1.4.csv"
cities = pd.read_csv(file)
cities = cities[["city_ascii", "state_name","state_id", "lat", "lng", "population"]]
cities_df = cities
cities_df
```

```python
results = []

for index, row in cities_df.iterrows():

    city = row["city_ascii"]
    state = row["state_name"]
    abbrev = row["state_id"]

    term = "speakeasy"
    location = f"{city}, {state}"
    categories = f"speakeasies,[{abbrev}]"
    sort_by = "best_match"
    limit = 50


    response = yelp_api.search_query(term = term, location = location, sort_by = sort_by, limit = limit)
    if(len(response['businesses']) == 0):
        continue
    df = pd.DataFrame([
        {
            'city' : city,
            'state' : state,
            'rating': b["rating"],
            'name': b["name"],
            "review_count": b["review_count"],
            "lat": b["coordinates"]["latitude"],
            "lng": b["coordinates"]["longitude"],
            "price" : b["price"] if("price" in b) else " "}
        for b in response["businesses"]
    ])

    results.append(df)
    
results_df = pd.concat(results)
```

```python
results_df.to_csv('speakeasies.csv', mode='w', header = False, index =False)
```
